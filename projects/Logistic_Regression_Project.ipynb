```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
# Reading training and testing data
train_data = pd.read_csv('Datasets/Logistic_Regression_train.csv',index_col = False)
test_data = pd.read_csv('Datasets/Logistic_Regression_test.csv',index_col = False)
```

# Preprocessing Training data

```python
train_data.drop("Name",axis = 1, inplace = True)
train_data
```

```python
train_data.drop("Ticket",axis = 1, inplace = True)
train_data
```

```python
train_data.drop("Cabin",axis = 1, inplace = True)
train_data
```

```python
temp=train_data.Sex
for i in range(len(temp)):
    if(temp[i]=='male'):
        temp[i]=1
    else:
        temp[i]=0
```

```python
train_data
```

```python
temp=train_data.Embarked
for i in range(len(temp)):
    if(temp[i]=='S'):
        temp[i]=0
    elif temp[i]=='C':
        temp[i]=1
    else: 
        temp[i]=2
```

```python
train_data
```

```python
train_data.Age.fillna(train_data.Age.mean(),inplace = True)
```

```python
train_data
```

```python
Y = train_data.pop('Survived')
X = train_data[:]
```
