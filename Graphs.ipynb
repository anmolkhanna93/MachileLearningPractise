# Customizing Graphs

```python
x1 = np.arange(0,5,0.1)
y = x1**3
plt.plot(x1,y,color = "black")
plt.show()
```

```python
x1 = np.arange(0,5,0.1)
y = x1**3
plt.plot(x1,y,color = "black",marker="o") # marker gives the points along with the line unlike "ro"
plt.show()
```

```python
x1 = np.arange(0,5,0.1)
y = x1**3
plt.plot(x1,y,color = "black",linewidth = 5)
plt.show()
```

```python
x1 = np.arange(0,5,0.1)
y = x1**3
y2 = x1 ** 2
plt.plot(x1,y,color = "red",marker="o",label = "X^3") # marker gives the points along with the line unlike "ro"
plt.plot(x1,y2,color = "blue",marker="p",label = "X^2")
plt.axis([0,10,0,150]) # 4 enteries
plt.grid()
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Demo Graph")
plt.legend() # required for us to show the legends we passed in plt.plot() function
plt.show()
```

```python
# Adding random text at some place in graph
x1 = np.arange(0,5,0.1)
y = x1**3
y2 = x1 ** 2
plt.plot(x1,y,color = "red",marker="o",label = "X^3") # marker gives the points along with the line unlike "ro"
plt.plot(x1,y2,color = "blue",marker="p",label = "X^2")
plt.axis([0,10,0,150]) # 4 enteries
plt.grid()
plt.ylabel("Y")
plt.xlabel("X")
plt.title("Demo Graph")
plt.legend() # required for us to show the legends we passed in plt.plot() function
plt.text(2,80,"Demo Random Text",fontsize = 30) 
plt.show()
```
