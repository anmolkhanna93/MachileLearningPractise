```python
import matplotlib.pyplot as plt
```

```python
x = [1,2,3]
y = [2,4,6]

plt.scatter(x,y)
plt.show()
```

```python
x = [1,2,3]
y = [2,4,6]
plt.scatter(x,y)
plt.plot(x,y)
plt.show()
```

```python
# Plotting function x ^3
import numpy as np
x = np.array([1,2,3])
y = x**3
plt.plot(x,y)
plt.show()
```

```python
x1 = np.arange(0,5,0.1)
y = x1**3
plt.plot(x1,y)
plt.show()
```

```python
x1 = np.arange(0,5,0.1)
y = x1**3
plt.plot(x1,y,"ro")
plt.show()
```
