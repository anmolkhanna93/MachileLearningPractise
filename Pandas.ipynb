# Manipulating data

```python
# deleting some rows. Here we are droping by labels
a = df.drop(0) #doing just df.drop() wont reflect the changes as drop function makes a new copy of the dataset with the 0th row removes
a.head()
# to reflect changes in df itself
df.drop(0,inplace = True)
df.head()
```

```python
df.drop(0,inplace = True)
df.head() 
# we get the following error because, rows were numbered from 0,1,2,3---- but after removing 0th row, indexing becomes 
# 1,2,3,4-----
```

```python
df.drop(3,inplace = True)
df.head() 
# here labels 0 and 3 are missing
# drop functions takes in labels and not positions
```

```python
#seeing the labels of our rows
df.index
```

```python
# droping by position
#df.index(3) gives label for row with position 3 and not label 3
df.drop(df.index[3],inplace = True)
df.head()
# we can also delete multiple rows also eg df.drop(df.index[[1,2]],inplace = True)
```

```python
df.SepalLength > 5 # number of rows with value greater than 5
```

```python
df[df.SepalLength > 5] # only rows with SepalLength >5
```

```python
df[df.flower_type=='Iris-setosa'].describe()
# we can create 3 classes like c1 = df[df.flower_type=='Iris-setosa'] and similarly c2 and c3 and comapre sepal lengths
```

```python
# iloc vs loc
# iloc works by position
print(df.iloc[0]) # 0th position row
# loc works by label
print(df.loc[10]) # row with label 10
```

```python
# adding a row
df.loc[0] = ["1","2","3","4","Iris-setosa"] # added at tail
```

```python
df.head() 
```

```python
df.tail()
```

```python
# reset index,needed because when ever we delete any row, index values become unordered
df.reset_index()
```

```python
# it does reset the indexes, but it has some issues
# Issue 1. created a new column index having previous index values, to rectify this, we do
df.reset_index(drop=True)
```

```python
# Issue 2: This reset_index() doesnt make changes in original dataset but gives you a new copy of it. So to get original 
# dataset,we do the following 
df.reset_index(drop=True,inplace = True)
df.index
```

```python
# Delete column
df.drop("SepalLength")
```

```python
# we get an error because its trying to find a label SepalLength, to rectify this
df.drop("SepalLength",axis = 1)# axis 1 means column wise
```

```python
df.describe() # here we are getting SepalLength because we didnt do inplace = True in above command so we got an updated copy
# and no changes were reflected in df
```
