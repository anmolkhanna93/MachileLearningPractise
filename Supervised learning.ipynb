```python
from sklearn import datasets
```

```python
boston = datasets.load_boston()
type(boston)
```

```python
print(boston)
```

```python
x = boston.data
y = boston.target
```

```python
type(x)
```

```python
x.shape # 506 rows and 13 columns i.e. 506 data points and 13 features
```

```python
# converting this data into the dataframe
import pandas as pd
df = pd.DataFrame(x)
df
```

```python
# seeing the features name
print(boston.feature_names)
```

```python
# setting the column/feature names of the data frame
df.columns = boston.feature_names
df
```

```python
# Deviding the data into testing and training data sets
from sklearn import model_selection
x_train, x_test, y_train, y_test= model_selection.train_test_split(x,y) # this function is very useful and important
```

```python
print(x_train.shape)
print(x_test.shape)
```

```python
# importing the linear regression algorithm
from sklearn.linear_model import LinearRegression
algorithm1 = LinearRegression()
```

```python
# training the algorithm
algorithm1.fit(x_train,y_train)
```

```python
# testing
y_predicted = algorithm1.predict(x_test)
```

```python
# comparing actual output with the predicted output
import matplotlib.pyplot as plt
plt.scatter(y_test,y_predicted) # predicted on y axis
plt.axis([0,40,0,40])
plt.show()
```
