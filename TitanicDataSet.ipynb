```python
# Deleting the name and ticket columns
del titanic['Name']
del titanic['Ticket']
```

```python
titanic.head()
```

```python
survived = titanic[titanic.Survived == 1]
```

```python
survived.reset_index(drop = True , inplace = True)
survived
```

```python
notsurvived = titanic[titanic.Survived == 0]
notsurvived.reset_index(drop = True , inplace = True)
```

```python
notsurvived
```

```python
m = (survived.Sex == 'male').sum()

f = (survived.Sex == 'female').sum()

l = [m,f]
labels = 'Male','Female'
explode = (0.1,0.1)
plt.pie(l,labels=labels , autopct='%1.1f%%' , explode=explode)
plt.title("SURVIVED")
plt.show()

m1 = (notsurvived.Sex == 'male').sum()
f1 = (notsurvived.Sex == 'female').sum()

l = [m1,f1]
labels = 'Male','Female'
explode = (0.1,0.1)
colors = ['Red','Blue']
plt.pie(l,labels=labels , autopct='%1.1f%%' , explode=explode , colors = colors)
plt.title("NOT - SURVIVED")
plt.show()
```
